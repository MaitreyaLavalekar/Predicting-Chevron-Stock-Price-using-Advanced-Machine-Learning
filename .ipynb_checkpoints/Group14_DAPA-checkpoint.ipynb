### Performing Normalization using Min / Max method

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Normalize the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)

# Split data into training and testing sets (80/20 split)
train_size = int(len(scaled_data) * 0.8)
train, test = scaled_data[:train_size], scaled_data[train_size:]

# Convert to pandas DataFrame for better handling
train_df = pd.DataFrame(train, columns=df.columns)
test_df = pd.DataFrame(test, columns=df.columns)
```

```python
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        a = data[i:(i + time_step), 3]  # Close prices
        X.append(a)
        y.append(data[i + time_step, 3])
    return np.array(X), np.array(y)

time_step = 30
X_train, y_train = create_dataset(train, time_step)
X_test, y_test = create_dataset(test, time_step)

# Reshape input to be [samples, time steps, features]
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Create the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=64, verbose=1)
```

```python
# Make predictions
y_pred = model.predict(X_test)

# Create an array of zeros with the same shape as the test data
# This will hold the inverse-transformed predictions
y_pred_full = np.zeros((len(y_pred), df.shape[1]))

# Since only the Close price was used for training, place the predictions in the correct column
y_pred_full[:, 3] = y_pred.flatten()

# Inverse transform the predictions
y_pred = scaler.inverse_transform(y_pred_full)[:, 3]

# Inverse transform the true values for comparison
y_test_full = np.zeros((len(y_test), df.shape[1]))
y_test_full[:, 3] = y_test.flatten()
y_test = scaler.inverse_transform(y_test_full)[:, 3]
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Evaluate
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'LSTM MSE: {mse}, MAE: {mae}, R^2: {r2}')
```

```python
# Plot actual vs predicted prices
plt.figure(figsize=(12, 6))

# Create a series for plotting purposes
actual_prices = pd.Series(y_test, index=df.index[train_size + time_step + 1:])
predicted_prices = pd.Series(y_pred, index=df.index[train_size + time_step + 1:])

# Plot actual prices
plt.plot(actual_prices, label='Actual Prices', color='Blue')

# Plot predictions
plt.plot(predicted_prices, label='LSTM Predictions', color = 'Red')

plt.title('Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
# Calculate profit or loss
profit_loss = y_pred - y_test

# Plot actual vs predicted prices and profit/loss
plt.figure(figsize=(14, 7))

# Plot actual prices
plt.plot(df.index[train_size + time_step + 1:], y_test, label='Actual Prices', color='blue')

# Plot predicted prices
plt.plot(df.index[train_size + time_step + 1:], y_pred, label='Predicted Prices', color='red')

# Highlight profit and loss areas
profit_indices = df.index[train_size + time_step + 1:][profit_loss > 0]
loss_indices = df.index[train_size + time_step + 1:][profit_loss < 0]

plt.scatter(profit_indices, y_test[profit_loss > 0], marker='o', color='green', label='Profit')
plt.scatter(loss_indices, y_test[profit_loss < 0], marker='o', color='orange', label='Loss')

plt.title('Chevron Stock Price Prediction with Profit and Loss Highlighted')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
