Calculating Daily Returns

Daily Return = (Closing Price on Day n − Closing Price on Day (n-1))/Closing Price on Day (n-1)


```python
# Calculate Daily Returns
data_frame['Daily Return'] = data_frame['Close'].pct_change()

#Printing the Daily Return percentage
print(data_frame['Daily Return'])

# Plot Daily Returns
plt.figure(figsize=(12, 6))
plt.plot(data_frame.index, data_frame['Daily Return'], label='Daily Return', color='purple')
plt.title('Daily Returns of Chevron Stock')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.grid(True)
plt.show()
```

#The above graph shows the perecentage of change in the closing pirce with respect to the previous day

```python
# Plot histogram of daily returns
plt.figure(figsize=(10, 6))
plt.hist(data_frame['Daily Return'].dropna(), bins=50, edgecolor='k')
plt.title('Histogram of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.show()
```

```python
# Add a 'Year' column for grouping
data_frame['Year'] = data_frame.index.year

# Plot histogram of daily returns, differentiated by year
plt.figure(figsize=(14, 8))

# Define a color palette
colors = plt.cm.viridis(np.linspace(0, 1, data_frame['Year'].nunique()))

# Plot histogram for each year
for i, year in enumerate(sorted(data_frame['Year'].unique())):
    yearly_data = data_frame[data_frame['Year'] == year]['Daily Return'].dropna()
    plt.hist(yearly_data, bins=50, alpha=0.5, color=colors[i], label=str(year))

plt.title('Histogram of Daily Returns Differentiated by Year')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.legend(title='Year')
plt.show()

```

```python
# Scatter plot of Volume vs. Closing Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_frame['Volume'], y=data_frame['Close'])
plt.title('Volume vs. Closing Price')
plt.xlabel('Volume')
plt.ylabel('Closing Price')
plt.show()
```

```python
# Calculate correlation matrix including the closing price
correlation_matrix = data_frame[['Open', 'High', 'Low', 'Close', 'Volume', 'Daily Return']].corr()

# Plot correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Stock Indicators')
plt.show()
```
